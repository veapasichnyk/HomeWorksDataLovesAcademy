1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# Символьне визначення змінної
x = sp.Symbol('x')

# Функція f(x)
f = 10 * x - 5

# Обчислення похідної
f_prime = sp.diff(f, x)
print(f"Похідна функції: f'(x) = {f_prime}")
```

```python
# Створюємо числові функції
f_lambdified = sp.lambdify(x, f, 'numpy')

# Діапазон значень x
x_vals = np.linspace(0, 10, 200)

# Значення функції
y_vals = f_lambdified(x_vals)

# Значення похідної (масив з тих самих значень)
y_prime_vals = np.full_like(x_vals, float(f_prime))  # Масив, заповнений 10
```

```python
# Побудова графіків
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x) = 10x - 5', color='blue')
plt.plot(x_vals, y_prime_vals, label="f'(x) = 10", color='red')
plt.title('Графік функції та її похідної')
plt.xlabel('Кількість годин навчання (x)')
plt.ylabel('Оцінка')
plt.grid(True)
plt.legend()
plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
# Символьна змінна
x = sp.symbols('x')

# Функції
f1 = sp.sin(x)
f2 = sp.ln(x)

# Похідні
df1 = sp.diff(f1, x)
df2 = sp.diff(f2, x)

# Вивід похідних
print("Похідна sin(x):", df1)
print("Похідна ln(x):", df2)
```

```python
# Перетворення в числові функції для побудови графіків
f1_func = sp.lambdify(x, f1, 'numpy')
df1_func = sp.lambdify(x, df1, 'numpy')

f2_func = sp.lambdify(x, f2, 'numpy')
df2_func = sp.lambdify(x, df2, 'numpy')
```

```python
# Точки для побудови графіків
x_vals1 = np.linspace(-2*np.pi, 2*np.pi, 400)
x_vals2 = np.linspace(0.1, 10, 400)  # для ln(x), щоб уникнути логарифма від 0
```

```python
# Побудова графіків
plt.figure(figsize=(12, 5))

# sin(x) та його похідна
plt.subplot(1, 2, 1)
plt.plot(x_vals1, f1_func(x_vals1), label='sin(x)', color='blue')
plt.plot(x_vals1, df1_func(x_vals1), label="d/dx sin(x)", color='red')
plt.title('sin(x) та його похідна')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# ln(x) та його похідна
plt.subplot(1, 2, 2)
plt.plot(x_vals2, f2_func(x_vals2), label='ln(x)', color='green')
plt.plot(x_vals2, df2_func(x_vals2), label="d/dx ln(x)", color='orange', linestyle='-')
plt.title('ln(x) та його похідна')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
# Символьна змінна
x = sp.symbols('x')

# Функція витрат
f = x**2 - 5*x + 1

# Похідна
df = sp.diff(f, x)
print("Похідна функції:", df)
```

```python
# Перетворення на числові функції
f_func = sp.lambdify(x, f, 'numpy')
df_func = sp.lambdify(x, df, 'numpy')
```

```python
# Значення x від 0 до 100
x_vals = np.linspace(0, 100, 500)
```

```python
# Побудова графіків
plt.figure(figsize=(10, 6))
plt.plot(x_vals, f_func(x_vals), label='f(x) = x² - 5x + 1', color='blue')
plt.plot(x_vals, df_func(x_vals), label="f'(x) = 2x - 5", color='red')

plt.title('Функція витрат і її похідна')
plt.xlabel('Кількість вироблених одиниць (x)')
plt.ylabel('Значення')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
